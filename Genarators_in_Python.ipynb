**Generators in Python**

Generators allows you to iterate over the data without storing the entire dataset in memory.

Instead of using the return keyword, they use yield keyword.

If the keyword yield is present means it is a generator

The code is just compiled in the generator, it is not executed, it will execute only when we iterate over the generator.

```python
def my_generator():
  yield 1
  yield 2

for value in my_generator():
  print(value)
```

```python
def count_upto(n):
  count =1
  while count <= n:
    yield count
    count += 1
for num in count_upto(5):
  print(num)

```

Unlike return, which terminates, yield pauses it and can be called multiple times
