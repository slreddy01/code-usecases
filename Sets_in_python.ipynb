SETS in Python

Sets are mutable, unordered, non duplicated .

Like list and Tuple this is mutable.

this is represented in {}

```python
fruits = {'apple','mango'}
fruits.add('grapes')
print(fruits)
```

frozenset keyword used for immutable set , we cant add or delete the values in set .

```python
animals=frozenset(['tiger','bear'])
animals
animals.add('cat')
animals
```

set operations

1.Membership
2.add
3.union
4.remove
5.clear
6.Intersection
7.Difference
8.SYM.Difference
9.Size
10.Copy

```python
#membership operation
set1 ={1,2,3}
6 in set1
```

```python
set1= {1,6,9}
6 in set1
```

```python
#remove operation

set1 ={ 6,9,10,12}
set1.remove(9)
set1
```

```python
#union operation
# if duplicates are there in both the sets, only single value will be copied

set1 ={8,10,11,17,19}
set2={1,2,19}

set1|set2
```

```python
#clear

set1.clear()
set1
```

```python
#intersection operation  will gives common values between 2 or more sets
#symbol for intersection is &
A = {1,2,3,4,5}
B={1,2,9,10}
A & B
```

```python
#minus symbol
#A-B means will give results other than common elements which are in A but not in B


A-B
```

```python
#if it is B-A

B-A
```

```python
#sym.difference symbol is ^
#will gives results which are not common elements in both sets
A^B
```

```python
#size and len() are same
len(A)
```

```python
#copy operation
c= {5,11,12}
A=c.copy()
print(A)
```
